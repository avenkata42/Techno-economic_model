```python
rate = [2000, 10000, 20000, 100000, 200000, 500000, 1000000, 2000000]
msp_scaled = [79.074, 62.874, 57.94, 49.438, 46.672, 43.626, 41.7, 40.042]
msp_unscaled = [88.668,
72.322,
67.314,
58.812,
56.046,
53,
51.072,
49.414
]

capex_scaled = [20.64604274,
101.6420565,
201.695956,
1008.47978,
2016.95956,
5042.398899,
10083.20964,
20164.83112
]

capex_unscaled = [67.10720812,
330.3739477,
655.5858024,
3277.929012,
6555.858024,
16389.64506,
32774.12803,
65543.09396
]

bop_scaled = [11.11709993,
54.73033814,
108.6055147,
543.0275737,
1086.055147,
2715.137869,
5429.420576,
10857.98599
]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

tot_scaled = [x+y for x,y in zip(capex_scaled,bop_scaled)]
tot_unscaled = [x+y for x,y in zip(capex_unscaled,bop_scaled)]

source = ['Electrolyzer cost', 'Balance of plant']

pct_max = 70 # for example, max percentile of color ramp
pct_min = 30 # for example, min percentile of color ramp
ramp = plt.cm.Blues # for example

plt.rcParams.update({'font.size': 20})
colorf = 'blue'

# number of items in data source
n = len(source)

# list of values between 0.00 and 1.00; length equals length of data source
n_prop = list(i / 100.0 for i in (np.arange(pct_min, pct_max, (pct_max-pct_min)/n)))

# create list of colors
clr_lst = []
for j in n_prop:
    clr = ramp(j) 
    clr_lst.append(clr)


fig, ax = plt.subplots(figsize =(6,6))
ax2 = ax.twinx()

ax.stackplot(rate, capex_unscaled, bop_scaled, labels=source, colors =clr_lst)
ax.stackplot(rate, capex_scaled, bop_scaled, labels=source, colors =clr_lst)
ax.margins(x=0)
#ax.legend(fontsize=4)
ax.set_xscale('log')
#ax.legend(loc='center left')
# ax.set_yscale('log')

ax.plot(rate, capex_scaled, ls = '--', linewidth =2, c='black')
ax.plot(rate, tot_scaled, ls = '--', linewidth =2, c='black')
ax.plot(rate, capex_unscaled, linewidth =0.75, c='black')
ax.plot(rate, tot_unscaled, linewidth =0.75, c='black')

ax2.plot(rate, msp_scaled, ls = '--', linewidth =2, c=colorf)
ax2.plot(rate, msp_unscaled, linewidth =1.5, c=colorf)

ax2.margins(x=0)
ax2.yaxis.label.set_color(colorf)
ax2.tick_params(axis='y', colors=colorf)

ax2.set_ylabel("Minimum Selling Price $C_{2}H_{4}$ (USD/kg)", c = colorf, fontsize =17)
ax2.spines['right'].set_color(colorf)

ax.set_xlabel("Ethylene Production rate (TPA)", fontsize =18)
ax.set_ylabel("Electrolyzer CAPEX(millions USD)", fontsize =18)
ax.tick_params(axis='x', labelsize=18)
ax.tick_params(axis='y', labelsize=18)
ax2.tick_params(axis='y', labelsize=18)
ax.legend(source, fontsize = 12)
plt.show()
```
